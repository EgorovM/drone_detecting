```python
import os

import numpy as np

import cv2
import matplotlib.pyplot as plt

from tqdm.notebook import tqdm
from IPython.display import clear_output


vid_cam1 = cv2.VideoCapture('video_1_cam_1.mp4')
vid_cam2 = cv2.VideoCapture('video_1_cam_2.mp4')


def get_video_frames(vidcap):
    success,image = vidcap.read()
    images = []
    
    for _ in tqdm(range(1300), desc='video_parsing', leave=False):  
        success, image = vidcap.read()
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        images.append(cv2.rotate(image, cv2.ROTATE_180))
        
    return images
```

```python
def get_frame_prefs(labels_dir, video_name, frame_number):
    file_path = os.path.join(labels_dir, 'labels', f"{video_name}_{frame_number}.txt")
    
    if not os.path.isfile(file_path):
        return [-1] * 6
    
    with open(file_path) as f:
        prefs = list(map(float, f.readline().split()))
        
    return prefs
```

```python
import numpy as np

ROOM_WIDTH = 6.815  # a
ROOM_LENGTH = 8.783  # b
ROOM_HEIGHT = 2.79

CAM_FLOOR_DEGREE = np.deg2rad(90 - 18.58)


IMG_1_PROP = {
    'dx': 660 - 800,  # room diagonal x pixel minus average pixel
    'dalpha': np.arctan(ROOM_LENGTH / ROOM_WIDTH),
    'dy': 308 - 448,  # room angle y pixel minus average pixel
    'width': 1920,
    'height': 896,
    'cam_degree': lambda: prefs['CAM_1_WALL_DEGREE'],
}

IMG_2_PROP = {
    'dx': 1107 - 960,  # room diagonal x pixel minus average pixel
    'dalpha': np.arctan(ROOM_LENGTH / ROOM_WIDTH),
    'width': 1920,
    'height': 1080,
    'cam_degree': lambda: prefs['CAM_2_WALL_DEGREE'],
}


def find_coords(alpha1, alpha2):
    x = ROOM_WIDTH * np.tan(alpha2) / (np.tan(alpha2) + np.tan(alpha1))
    y = x * np.tan(alpha1)
    
    return x, y


def find_alpha(bbox, img_prop):
    x = (1 - bbox[1] + bbox[3] // 2) * img_prop['width']
    x1 = img_prop['width'] / 2
    
    dalpha = (img_prop['dalpha'] - img_prop['cam_degree']()) / img_prop['dx'] * (x - x1)

    return img_prop['cam_degree']() + dalpha


def find_floor_degree(bbox, img_prop):
    y = (1 - bbox[2] + bbox[4] // 2) * img_prop['height']
    y1 = img_prop['height'] / 2
    
    dbeta = (np.arctan(ROOM_LENGTH / ROOM_HEIGHT) - CAM_FLOOR_DEGREE) / img_prop['dy'] * (y - y1)
    
    return CAM_FLOOR_DEGREE + dbeta 


def find_alpha1(bbox):
    return find_alpha(bbox, IMG_1_PROP)

def find_alpha2(bbox):
    return find_alpha(bbox, IMG_2_PROP)

def find_x_y_coord(bbox1, bbox2):
    alpha1 = find_alpha1(bbox1)
    alpha2 = find_alpha2(bbox2)
    
    coords = find_coords(alpha1, alpha2)
    
    return coords


def find_z_coord(bbox1, y):
    beta = find_floor_degree(bbox1, IMG_1_PROP)
    
    return (ROOM_HEIGHT * np.tan(beta) - y) / np.tan(beta)
```

```python
def bbox_to_yolo(bbox, h, w):
    return [
        0, 
        1 - bbox[0] / w, 
        1 - bbox[1] / h, 
        abs(bbox[2] - bbox[0]) / w,
        abs(bbox[3] - bbox[1]) / h,
        1
    ]
```

```python
import numpy as np
import cv2

def rotate_image(image, angle):
    image_center = tuple(map(int, np.array(image.shape[1::-1]) / 2))
    rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
    result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
    result[image_center[1]-10:image_center[1]+10, image_center[0]-10:image_center[0]+10] = [255, 0, 0]
    return result
```

```python
opt_deg1 = None
opt_deg2 = None

opt_mae = 10000

for deg1 in np.linspace(42, 52, 500):
    for deg2 in np.linspace(42, 52, 500):
        prefs['CAM_1_WALL_DEGREE'] = np.deg2rad(deg1)
        prefs['CAM_2_WALL_DEGREE'] = np.deg2rad(deg2)

        new_mae = 0
        
        for box_img in boxes_imgs:
            x, y = find_x_y_coord(box_img['prefs1'], box_img['prefs2'])
            x_real, y_real, _ = box_img['real_coords']
                
            new_mae += (x - x_real) ** 2 + abs(y - y_real) ** 2

        new_mae = new_mae / 4

        if new_mae < opt_mae:
            opt_mae = new_mae
            opt_deg1 = deg1
            opt_deg2 = deg2
```
