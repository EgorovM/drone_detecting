```python
import cv2
import numpy as np
from functools import lru_cache
```

```python
#Optional
import matplotlib.pyplot as plt
import plotly.express as pe
```

```python
class Camera:
    def __init__(self, param_path, output_height, output_width):
        with open(param_path) as f:
            param_dict = eval(f.read())
            
            for key, val in param_dict.items():
                setattr(self, key, val)
                
        self.output_height = output_height
        self.output_width = output_width
    
    @lru_cache
    def get_newcameramatrix(self):
        h, w = self.output_height, self.output_width
        newcameramtx, _ = cv2.getOptimalNewCameraMatrix(self.mtx, self.dist, (w, h), 1, (w, h))
        
        return newcameramtx

    def transform_image(self, img):
        newcameramtx = self.get_newcameramatrix()
        return cv2.undistort(img, self.mtx, self.dist, None, newcameramtx)
    
    def transform_point(self, point):
        newcameramtx = self.get_newcameramatrix()
        
        return np.round(cv2.undistortPoints(
            point,
            self.mtx,
            self.dist, 
            None, 
            None, 
            newcameramtx
        )[0][0]).astype(int)
    
    def localize(self, x, y):
        pass
    
    def read_img(self, img_path):
        img = cv2.imread(img_path)
        img = self.transform_image(img)
        
        return img
```

```python
class Room:
    def __init__(self, cam1, cam2, length, width, height):
        self.cam1 = cam1
        self.cam2 = cam2
        self.length = length
        self.width = width
        self.height = height
        
class RoomObject:
    def __init__(self, room, x, y, z):
        self.x, self.y, self.z = x, y, z
        self.room = room
    
    def _detect(self):
        pass
    
    def detect(self):
        cam1 = self.room.cam1
        cam2 = self.room.cam2
        
        bbox1, bbox2 = self._detect()
        bbox1 = [*cam1.transform_point(bbox1[:2]), *cam1.transform_point(bbox1[2:])]
        bbox2 = [*cam2.transform_point(bbox2[:2]), *cam2.transform_point(bbox2[2:])]
        
        return bbox1, bbox2
    
    def localize(self):
        bbox1, bbox2 = self.detect()
        
def _replace_coordinates(room, x, y, z):
    """ Переводим координаты из изображения в собственные координаты """ 
    return y, room.length - x, z 

def _create_room_object(room, x, y, z):
    x, y, z = _replace_coordinates(room, x, y, z)
    
    return RoomObject(room, x, y, z)
```

```python
class Localization:
    def __init__(self, room, cam1, cam2):
        self.room = room
        self.cam1 = cam1
        self.cam2 = cam2
        
    def find_coords(self, alpha1, alpha2):
        x = self.room.width * np.tan(alpha2) / (np.tan(alpha2) + np.tan(alpha1))
        y = x * np.tan(alpha1)

        return x, y

    def find_alpha(self, bbox, cam):
        x = ((bbox[0] + bbox[2]) // 2)
        x1 = cam.output_width / 2
        
        dalpha = np.arctan(self.room.length / self.room.width)        
        dalpha = (dalpha - cam.wall_degree) / cam.dx * (x - x1)

        return cam.wall_degree + dalpha

    def find_floor_degree(self, bbox, cam):
        y = ((bbox[1] + bbox[3]) // 2)
        y1 = cam.output_height / 2

        dbeta = np.arctan(self.room.length / self.room.width)
        dbeta = (dbeta - cam.floor_degree) / cam.dy * (y - y1)

        return cam.floor_degree + dbeta 

    def find_x_y_coord(self, bbox1, bbox2):
        alpha1 = self.find_alpha(bbox1, self.cam1)
        alpha2 = self.find_alpha(bbox2, self.cam2)
        
        # print(np.rad2deg(alpha1), np.rad2deg(alpha2))
        
        coords = self.find_coords(alpha1, alpha2)
        
        return coords

    def find_z_coord(self, bbox1, y, cam=None):
        if not cam:
            cam = self.cam1
            
        beta = self.find_floor_degree(bbox1, cam)
        return (self.room.height * np.tan(beta) - y) / np.tan(beta)
    
    def localize(self, bbox1, bbox2):
        x, y = self.find_x_y_coord(bbox1, bbox2)
        
        z1 = self.find_z_coord(bbox1, x, self.cam1)
        z2 = self.find_z_coord(bbox2, x, self.cam2)
        z = (z1 + z2) / 2
        
        return x, y, z
```

```python
cam1 = Camera('camera_parameters/cam1_params.txt', 896, 1600)
cam2 = Camera('camera_parameters/cam2_params.txt', 1920, 1920)

room = Room(cam1, cam2, 8.783, 6.815, 2.79)

room_objects = [
    _create_room_object(room, 0.446, 3.314, 0.75),
    _create_room_object(room, 4.298, 5.437, 0.768),
    _create_room_object(room, 4.861, 3.416, 0.747),
    _create_room_object(room, 6.250, 2.623, 0.977),
]

# simulating the detection
bboxes_list = (
    [(277, 307, 318, 327),  (1429, 264, 1473, 284)],
    [(994, 307, 1018, 343), (1601, 564, 1665, 635)],
    [(774, 455, 831, 484),  (904, 534, 973, 573)],
    [(892, 590, 969, 643),  (384, 598, 462, 642)],
)

for room_obj, bboxes in zip(room_objects, bboxes_list):
    room_obj._detect = lambda bboxes=bboxes: bboxes
```

```python
localizator = Localization(room, cam1, cam2)
```

```python
for room_obj in room_objects:
    bbox1, bbox2 = room_obj.detect()
    x, y, z = localizator.localize(bbox1, bbox2)
    
    print(x, y, z)
    # print(x - room_obj.x, y - room_obj.y, z - room_obj.z)
```

```python
def draw_bbox(img, bbox1):
    x1, y1, x2, y2 = bbox1
    
    img = img.copy()
    cv2.rectangle(img, (x1, y1), (x2, y2), (255,0,0), 2)
    plt.figure(figsize=(16, 9))
    plt.imshow(img)
```
